```python
def compute_final_metrics(data: dict, k: int) -> tuple[int, float, float]:
    """
    Given a `data` dictionary, computes the number of scans, overall nodule recall,
    the mean number of unmatched predicted nodules per scan for scans that have at most
    `k` true nodules.
    """

    results = {
        "total_true_nodules": 0,
        "total_matched_nodules": 0,
        "total_unmatched_predicted_nodules": 0,
        "scan_stats": {}  # per scan detail
    }

    for scan_id, scan_data in data.items():
        num_true = scan_data["num_true_volumes"]

        if num_true <= k:
            num_matched = len([val for val in scan_data["nodule_identification"] if val])
            num_unmatched = scan_data["num_unmatched_pred_volumes"]

            results["total_true_nodules"] += num_true
            results["total_matched_nodules"] += num_matched
            results["total_unmatched_predicted_nodules"] += num_unmatched

            assert num_true >= 0
            assert num_unmatched >= 0
            assert num_matched >= 0

            results["scan_stats"][scan_id] = {
                "true_nodules": num_true,
                "matched_nodules": num_matched,
                "unmatched_predicted": num_unmatched,
                "recall": num_matched / num_true if num_true > 0 else None,
                "predicted_vols": [entry["predicted_volume"] for entry in scan_data["matched_volumes"]],
            }
    
    # plot_predicted_volumes(results["scan_stats"])

    overall_recall = results["total_matched_nodules"] / results["total_true_nodules"]
    overall_precision = results["total_matched_nodules"] / (results["total_matched_nodules"] + results["total_unmatched_predicted_nodules"])
    mean_number_of_unmatched_predicted_nodules_per_scan = np.mean([results["scan_stats"][scan_id]["unmatched_predicted"] for scan_id in results["scan_stats"]])
    median_number_of_unmatched_predicted_nodules_per_scan = np.median([results["scan_stats"][scan_id]["unmatched_predicted"] for scan_id in results["scan_stats"]])

    print(f"=== Sensitivity & Specificity Analysis (k = {k if k != float('inf') else 'all'}, N = {len(results['scan_stats'])}) ===")
    print(f"Total True Nodules: {results['total_true_nodules']}")
    print(f"Total Matched Nodules (True Positives): {results['total_matched_nodules']}")
    print(f"Total Unmatched Predicted Nodules (False Positives): {results['total_unmatched_predicted_nodules']}")
    print(f"\n>> Overall Recall (Sensitivity): {overall_recall:.4f}")
    print(f">> Overall Precision (Specificity): {overall_precision:.4f}")
    print(f">> Mean Number of Unmatched Predicted Nodules per Scan: {mean_number_of_unmatched_predicted_nodules_per_scan:.4f}")
    print(f">> F1 Score: {((1 + (1**2)) * ((overall_precision * overall_recall) / (((1**2) * overall_precision) + overall_recall))):.4f}")
    print(f">> F4 Score: {((1 + (4**2)) * ((overall_precision * overall_recall) / (((4**2) * overall_precision) + overall_recall))):.4f}")
    print(f">> Median Number of Unmatched Predicted Nodules per Scan: {median_number_of_unmatched_predicted_nodules_per_scan:.4f}\n")
    print("\n")

    return overall_recall, mean_number_of_unmatched_predicted_nodules_per_scan, median_number_of_unmatched_predicted_nodules_per_scan
```

```python
def compute_recall_by_volume(data: dict, max_nodule_volume: float, k: int) -> float:
    """
    Computes overall recall (sensitivity) for true nodules with volume <= `max_nodule_volume`.
    Ignores unmatched predicted nodules.

    Only accounts for scans with `k` or fewer true nodules.
    """

    total_true = 0
    total_matched = 0

    for scan_data in data.values():
        if scan_data["num_true_volumes"] <= k:
            for entry in scan_data["matched_volumes"]:
                if entry["true_volume"] <= max_nodule_volume:
                    if entry["predicted_volume"] > 0: # actually has a match!
                        total_matched += 1
                    total_true += 1

    recall = total_matched / total_true if total_true > 0 else 0.0

    print(f"=== Recall Analysis (Nodule Volume ≤ {max_nodule_volume}) ===")
    print(f"Total Qualified True Nodules: {total_true}")
    print(f"Total Matched Nodules (True Positives): {total_matched}")
    print(f">> Overall Recall (Sensitivity): {recall:.4f}\n")

    return recall
```

```python
def plot_recall_and_unmatched_vs_k(k_to_metrics):
    # Separate numeric ks and the 'all' key
    numeric_ks = sorted([k for k in k_to_metrics if k != "all"])
    all_k = "all" if "all" in k_to_metrics else None

    # X values for plotting
    x_vals = numeric_ks + ([all_k] if all_k else [])
    x_positions = list(range(len(x_vals)))
    xtick_labels = [str(k) if k != float("inf") else "all" for k in x_vals]

    # Y-axis data
    recalls = [k_to_metrics[k]["recall"] for k in x_vals]
    mean_unmatched = [k_to_metrics[k]["mean_unmatched_preds"] for k in x_vals]

    fig, ax1 = plt.subplots(figsize=(10, 6))

    color1 = 'tab:blue'
    ax1.set_xlabel('k (Max # of True Nodules in Scan)')
    ax1.set_ylabel('Recall (Sensitivity)', color=color1)
    recall_line, = ax1.plot(x_positions, recalls, color=color1, marker='o', label='Recall')
    ax1.tick_params(axis='y', labelcolor=color1)
    ax1.set_ylim(0, 1.05)

    # Annotate recall points (higher offset)
    for x, y in zip(x_positions, recalls):
        ax1.annotate(f"{y:.2f}", (x, y), textcoords="offset points", xytext=(0, 12), ha='center', color=color1)

    # Right y-axis
    ax2 = ax1.twinx()
    color2 = 'tab:red'
    ax2.set_ylabel('Mean # of Unmatched Predicted Nodules per Scan', color=color2)
    unmatched_line, = ax2.plot(x_positions, mean_unmatched, color=color2, marker='s', linestyle='--', label='Unmatched Predictions')
    ax2.tick_params(axis='y', labelcolor=color2)

    # Annotate unmatched predictions (lower offset)
    for x, y in zip(x_positions, mean_unmatched):
        ax2.annotate(f"{y:.2f}", (x, y), textcoords="offset points", xytext=(0, -18), ha='center', color=color2)

    # X ticks
    plt.xticks(x_positions, xtick_labels)
    plt.title('Recall and Mean Unmatched Predictions Across k')
    fig.tight_layout()
    plt.grid(True)
    plt.show()
```

```python
def plot_predicted_volumes(data):
    all_predicted_vols = []
    for patient_data in data.values():
        all_predicted_vols.extend(patient_data["predicted_vols"])
    
    mean_volume = np.mean(all_predicted_vols)
    median_volume = np.median(all_predicted_vols)

    plt.figure(figsize=(10, 6))
    plt.hist(all_predicted_vols, bins=20, edgecolor='black', alpha=0.7)
    plt.axvline(mean_volume, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_volume:.1f}')
    plt.axvline(median_volume, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_volume:.1f}')

    plt.xlabel('Predicted Volume')
    plt.ylabel('Count')
    plt.title('Distribution of Matched Predicted Volumes')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_histogram(values, bins=50, title='Histogram', xlabel='Value', ylabel='Frequency'):
    mean_value = np.mean(values)
    median_value = np.median(values)

    plt.hist(values, bins=np.linspace(0, 1, bins), edgecolor='black', alpha=0.7)
    plt.title(f"{title} [N = {len(values)}]")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    text_str = f'Mean: {mean_value:.4f}\nMedian: {median_value:.4f}'
    plt.text(0.95, 0.95, text_str, transform=plt.gca().transAxes, fontsize=10,
             verticalalignment='top', horizontalalignment='right', bbox=dict(facecolor='white', alpha=0.5))
    
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
